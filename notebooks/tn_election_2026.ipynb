## Extract Data From ECI - using Basic Selenium Scraper

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()

url = "https://results.eci.gov.in/ResultAcGenMay2026/"
driver.get(url)

time.sleep(10)

# Find first table
table = driver.find_element(By.TAG_NAME, "table")

# Extract rows
rows = table.find_elements(By.TAG_NAME, "tr")

data = []

for row in rows:

    cols = row.find_elements(By.TAG_NAME, "td")

    # Handle header row
    if not cols:
        cols = row.find_elements(By.TAG_NAME, "th")

    row_data = [col.text for col in cols]

    data.append(row_data)

# Convert to DataFrame
df = pd.DataFrame(data[1:], columns=data[0])

print(df.head())

df.to_csv("eci_results_clean.csv", index=False)

driver.quit()
```

## Scrapping ECI Statistical Reports

```python
import pdfplumber
import pandas as pd

all_tables = []

with pdfplumber.open("10-Detailed_Results_1778165153.pdf") as pdf:
    for page_num, page in enumerate(pdf.pages):

        tables = page.extract_tables()

        for table_num, table in enumerate(tables):

            df = pd.DataFrame(table[1:], columns=table[0])

            df.to_csv(
                f"page_{page_num+1}_table_{table_num+1}.csv",
                index=False
            )

print("Extraction completed")
```

```python
import pdfplumber
import pandas as pd
import re

all_rows = []

pdf_file = "10-Detailed_Results_1778165153.pdf"

with pdfplumber.open(pdf_file) as pdf:

    current_constituency = ""

    for page in pdf.pages:

        text = page.extract_text()

        if not text:
            continue

        lines = text.split("\n")

        for line in lines:

            line = line.strip()

            # Detect constituency
            if "Constituency" in line:

                current_constituency = line

                continue

            # Skip unwanted lines
            skip_keywords = [
                "Election Commission",
                "DETAILED RESULTS",
                "TOTAL VALID",
                "CANDIDATE NAME",
                "Page ",
                "ST4895"
            ]

            if any(keyword in line for keyword in skip_keywords):
                continue

            # Remove multiple spaces
            line = re.sub(r"\s+", " ", line)

            # Candidate rows start with serial number
            match = re.match(
                r"^(\d+)\s+(.*?)\s+(MALE|FEMALE|OTHER|NOTA)\s+(\d+|NOTA)\s+(GENERAL|SC|ST|NOTA)\s+(.*?)\s+(.*?)\s+(\d+)\s+(\d+)\s+(\d+)\s+([\d.]+)\s+([\d.]+)$",
                line
            )

            if match:

                row = {
                    "Constituency": current_constituency,
                    "Serial_No": match.group(1),
                    "Candidate_Name": match.group(2),
                    "Gender": match.group(3),
                    "Age": match.group(4),
                    "Category": match.group(5),
                    "Party": match.group(6),
                    "Symbol": match.group(7),
                    "General_Votes": match.group(8),
                    "Postal_Votes": match.group(9),
                    "Total_Votes": match.group(10),
                    "Percent_Valid_Votes": match.group(11),
                    "Percent_Total_Electors": match.group(12)
                }

                all_rows.append(row)

df = pd.DataFrame(all_rows)

df.to_csv("TamilNadu_2026_Detailed_Results.csv", index=False)

print(df.head())

print(f"\nTotal Rows Extracted: {len(df)}")
```

## Scrapping of Perfomance of Political Parties 2026

```python
import camelot

tables = camelot.read_pdf(
    "Performance_of_Political_Parties_1778165153 (1).pdf",
    pages='all',
    flavor='stream'   # good for bordered tables
)

print("Tables found:", tables.n)
```

```python
for i, table in enumerate(tables):
    print(f"\n===== TABLE {i} =====")
    print(table.df.head(15))
```

```python
import pandas as pd

dfs = []

for i in [1, 2, 3]:
    dfs.append(tables[i].df)

combined_df = pd.concat(dfs, ignore_index=True)

print(combined_df.head(30))
```

```python
header1 = combined_df.iloc[0]
header2 = combined_df.iloc[1]

new_cols = []

for h1, h2 in zip(header1, header2):

    h1 = str(h1).strip()
    h2 = str(h2).strip()

    if h2 and h2 != 'nan':
        col = f"{h1}_{h2}"
    else:
        col = h1

    col = (
        col.lower()
        .replace(" ", "_")
        .replace("%", "pct")
        .replace("+", "plus")
        .replace("-", "_")
    )

    new_cols.append(col)

combined_df.columns = new_cols
```

```python
combined_df = combined_df.iloc[2:].reset_index(drop=True)
```

```python
print(combined_df.head(20))
print(combined_df.columns)
```

```python
combined_df = combined_df.rename(columns={
    combined_df.columns[0]: 'party',
    combined_df.columns[1]: 'candidates_contested',
    combined_df.columns[2]: 'candidates_won',
    combined_df.columns[3]: 'candidates_fd'
})
```

```python
combined_df.to_csv(
    "political_parties_2026_clean.csv",
    index=False
)

print("CSV exported successfully")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

url = "https://www.indiavotes.com/vidhan-sabha/2026/tamil-nadu/307/40"

# Launch browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open page
driver.get(url)

# Wait for JavaScript content
time.sleep(8)

# Get fully rendered HTML
html = driver.page_source

print(html[:2000])

# Extract tables
tables = pd.read_html(html)

print("Tables found:", len(tables))

# Save first table
df = tables[0]

print(df.head())

df.to_csv("indiavotes.csv", index=False)

driver.quit()
```

## Data Cleaning - TamilNadu-2026-Detailed-Results

```python
import pandas as pd

df = pd.read_csv("Tamilnadu_2026_Detailed_Results.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Extract data
extracted = df['Constituency'].str.extract(
    r'(\d+)-([^(]+)\((\d+)\)'
)

# Assign columns
extracted.columns = [
    'Constituency_No',
    'Constituency_Name',
    'Total_Electors'
]

# Merge back
df = pd.concat([df, extracted], axis=1)

# Clean types
df['Constituency_No'] = pd.to_numeric(
    df['Constituency_No'],
    errors='coerce'
)

df['Total_Electors'] = pd.to_numeric(
    df['Total_Electors'],
    errors='coerce'
)

df['Constituency_Name'] = (
    df['Constituency_Name']
    .str.strip()
)

print(df.head())
```
